```python
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
with open('../../Data/Pb_conc-predictions-no_coords.csv', mode='r', newline='') as csvfile:
    pb_conc_preds = pd.read_csv(csvfile)
```

```python
with open('../../Data/67Pb-predictions-no_coords.csv', mode='r', newline='') as csvfile:
    pb67_preds = pd.read_csv(csvfile)
```

```python
with open('../../Data/87Pb-predictions-no_coords.csv', mode='r', newline='') as csvfile:
    pb87_preds = pd.read_csv(csvfile)
```

```python
col_names = list(pb_conc_preds.columns)
col_names.remove('Pb_conc')
```

```python
all_Pb_preds = pb_conc_preds.merge(pb67_preds, on=col_names)
all_Pb_preds = all_Pb_preds.merge(pb87_preds, on=col_names)
all_Pb_preds
```

```python
all_Pb_preds_idxs = all_Pb_preds.set_index(['WOD_latitude [degrees_north]', 'WOD_longitude [degrees_east]', 'Depth'])
all_Pb_preds_xr = xr.Dataset.from_dataframe(all_Pb_preds_idxs)
all_Pb_preds_xr
```

```python
all_Pb_preds_xr.to_netcdf('../../Data/Pb_climatology.nc', format='NETCDF4')
```

```python
all_Pb_preds.to_csv('../../Data/Pb_climatology.csv', index=False)
```
